Polypmorphism

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

#Inheritance
class Student(Person):
  def traits(self):
    return f"I love cars, playing billiards, and watching F1"

# Define Teacher class OUTSIDE Person
class Teacher(Person):
  def traits(self):
    return f"i love teaching, socializing,"

# Create instances
student = Student("CJ", 19)
teacher = Teacher("Faye", 29)

# Call methods
print("These are my hobbies as student:" , student.traits())
print("These are my hobbies as a teacher:" , teacher.traits())
```

apply polymorphism

```python
class CarParts:
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Engine(CarParts):
  def traits(self):
    return f"An engine is some machine that converts energy from a fuel to some mechanical energy, creating motion in the process. "

class Transmission(CarParts):
  def traits(self):
    return f" acts as the medium that transmits power generated by the engine to the wheels via a mechanical system of gears and gear trains"

class SuperCharger(CarParts):
  def traits(self):
    return f"increases the pressure of air supplied to an internal combustion engine."

class Suspension(CarParts):
  def traits(self):
    return f"designed to help maximize the amount of friction between the tires and the road"

class Exhaust(CarParts):
  def traits(self):
    return f"The exhaust system collects the exhaust gases from the cylinders, removes harmful substances"


engine = Engine("engine", 10)
transmission = Transmission("transmission", 10)
supercharger = SuperCharger("supercharger", 10)
suspension = Suspension("suspension", 10)
exhaust = Exhaust("Exhaust", 10)

print("Engine Function:" , engine.traits())
print("Transmission Function:" , transmission.traits())
print("Supercharger Function:" , supercharger.traits())
print("Suspension Function:" , suspension.traits())
print("Exhaust Function:" , exhaust.traits())
```
